```python
#이 작업에서 중요한 부분 : dictionary의 key value"조립", list에 넣어"저장(추가)"

print("==CLI 게시판 실행==")
a_list = [] #while 밖에 있어야 루프 돌고 하나씩 저장(게시글 저장장소)
k = 0       #while 밖에 있어야 루프 돌고 하나씩 증가

while True: #명령어를 계속 받고 동작하기 위해
  cmd = input(" 명령어) ").strip()#cmd 입력시 공백 생겼을 때 없애주는 기능.strip()
  if cmd == "exit":
    break

  elif cmd == "article write":
    k += 1
    title = input("제목:")
    content = input("내용:")
    tw = {"번호": k,"제목": title,"내용": content} #dictionary의 key value"조립"
    #print(tw) #tw 잘 만들어졌는지 픞린트해서 내가 확인하기
    a_list.append(tw) #조립한거 a_list에 넣어"저장(추가)"
    print("{}번째 글이 작성되었습니다".format(k))

  elif cmd == "article list": #번호, 제목, 내용을 묶어서 저장 #dic, list이용해야하는데...
    if len(a_list) == 0:  # == if not a_list
      print("게시글이 없습니다")

    else:
      print("======================================")
      print("번호/", "제목/", "내용")
      for tw in reversed(a_list):
        print("{}/ {}/ {}".format(tw["번호"], tw["제목"], tw["내용"])) # "key"

      print("======================================")
  else:
    print("지원하지 않는 기능입니다")


print("==CLI 게시판 종료==")
```

```python
#이 작업에서 중요한 부분 : 작성글 삭제
print("==CLI 게시판 실행==")
a_list = []
k = 0

while True:
  cmd = input(" 명령어) ").strip()

  if cmd == "exit":
    break

  elif cmd == "article write":
    k += 1
    id = k #k번째이긴 하지만 사이클 번호랑 헷갈리면 안될듯하여 추가
    title = input("제목:")
    content = input("내용:")
    tw = {"번호": id,"제목": title,"내용": content} #k를 id 라는 변수로 수정
    a_list.append(tw) #조립한거 a_list에 넣어"저장(추가)"
    print("{}번째 글이 작성되었습니다".format(id)) # k->id

  elif cmd == "article list":
      if not a_list:
        print("게시글이 없습니다")

      else:
        print("======================================")
        print("번호/", "제목/", "내용")
        for tw in reversed(a_list):
          print("{}/ {}/ {}".format(tw["번호"], tw["제목"], tw["내용"])) # "key"
        print("======================================")


  elif cmd.startswith("article delete"):
    cmd_sp = cmd.split() #.split() 띄어뜨기(공백) 기준으로 자르겠다
    delete_id = int(cmd_sp[2])
    found = False

    ##입력을 "article delete 값"하면
    ##띄어쓰기 기준 잘라서 print하면 리스트로 담긴게 보임
    ### print하면 ["article", "delete", "값"]
    ##확인하고싶은 것은 리스트 내 인덱스 2인 "값"

    for tw in a_list:
      if tw["번호"] == delete_id:
            a_list.remove(tw)
            found = True
            print("{}번 글이 삭제되었습니다.".format(delete_id))
            break
    if found == False:
      print("{}번 글은 존재하지 않습니다.".format(delete_id))

  else:
    print("지원하지 않는 기능입니다")

print("==CLI 게시판 종료==")

# 해결할 사항
# 게시글 삭제: article delete ""
# 게시글 수정: article
# 게시글 상세보기: article
```
